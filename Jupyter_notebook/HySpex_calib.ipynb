```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
import itertools
import seaborn as sns
```

```python
palette = itertools.cycle(sns.color_palette())
sns.set_context("talk")
```

```python
roll_vnir = [1.237, 1.230, 1.224]
roll_vnir_err = [0.019, 0.012, 0.032]
pitch_vnir = [-0.284, -0.341, -0.305]
pitch_vnir_err = [0.019,  0.033, 0.019]
head_vnir = [0.051, -0.738, -0.421]
head_vnir_err = [0.060,  0.297, 0.102]
roll_swir = [-0.307, -0.268, -0.308]
roll_swir_err = [ 0.030, 0.009, 0.025]
pitch_swir = [0.191, 0.118 , 0.137]
pitch_swir_err = [0.033,  0.024, 0.025]
head_swir = [-0.160, -0.989, -0.670]
head_swir_err = [0.022, 0.128, 0.060]
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
fig.set_figwidth(18)
ax1.errorbar([1, 2, 3], roll_vnir, yerr=roll_vnir_err, fmt='o', label="roll");
c = next(palette)
ax2.errorbar([1, 2, 3], pitch_vnir, yerr=pitch_vnir_err, fmt='o', color="darkred", label="pitch");
c = next(palette)
ax3.errorbar([1, 2, 3], head_vnir, yerr=head_vnir_err, fmt='o', color="darkgreen", label="heading");
fig.suptitle("VNIR Boresight offsets")
for ax in [ax1,ax2,ax3]:
    ax.legend(loc="lower left")
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True)
fig.set_figwidth(15)
fig.set_figheight(15)
ax1.errorbar([1, 2, 3], roll_vnir, yerr=roll_vnir_err, fmt='o', label="roll");
c = next(palette)
ax2.errorbar([1, 2, 3], pitch_vnir, yerr=pitch_vnir_err, fmt='o', color="darkred", label="pitch");
c = next(palette)
ax3.errorbar([1, 2, 3], head_vnir, yerr=head_vnir_err, fmt='o', color="darkgreen", label="heading");
fig.suptitle("VNIR Boresight offsets")
for ax in [ax1,ax2, ax3]:
    ax.legend()
```

```python
plotoptions = {
    'fmt': 'o',
    'ecolor': 'lightgrey', 
    'elinewidth': 3, 
    'markersize': 6
}
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
fig.set_figwidth(13)
fig.set_figheight(9)
ax1.errorbar([1, 2, 3], roll_vnir, yerr=roll_vnir_err, label="roll", 
             color="tab:purple", **plotoptions)
ax1.errorbar([1, 2, 3], pitch_vnir, yerr=pitch_vnir_err, label="pitch",
            color="tab:green", **plotoptions)
ax1.errorbar([1, 2, 3], head_vnir, yerr=head_vnir_err, label="heading",
             color="tab:orange", **plotoptions)
ax1.legend(loc=(0.65,0.7), frameon=True, edgecolor="tab:grey")
ax1.set_title("VNIR camera offsets (deg)")
ax1.set_xticks((1,2,3))
ax1.set_xlim([.5,3.5])
ax1.set_xticklabels(("May 22", "Jun 25" , "Sep 12"))
ax1.axvline(x=1.5, ls='--', lw=2, c="grey")
ax1.set_ylabel("Degrees")
ax2.errorbar([1, 2, 3], roll_swir, yerr=roll_swir_err, label="roll", 
             color="tab:purple", **plotoptions)
ax2.errorbar([1, 2, 3], pitch_swir, yerr=pitch_swir_err, label="pitch",
            color="tab:green", **plotoptions)
ax2.errorbar([1, 2, 3], head_swir, yerr=head_swir_err, label="heading",
             color="tab:orange", **plotoptions)
ax2.legend(loc=(0.65,0.7), frameon=True, edgecolor="tab:grey")
ax2.set_title("SWIR camera offsets (deg)")
ax2.set_xticks((1,2,3))
ax2.set_xlim([.5,3.5])
ax2.set_xticklabels(("May 22", "Jun 25" , "Sep 12"))
ax2.axvline(x=1.5, ls='--', lw=2, c="grey")
fig.tight_layout()
```
